```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import re
from datetime import datetime
```

### 파일 불러오기

```python
path = os.path.join(root, 'data', 'raw')
files = [f for f in listdir(path) if isfile(join(path, f))]

seoul_routes = []
for file in files:
    filepath = os.path.join(path, file)
    route = pd.read_csv(filepath, encoding='CP949')
    seoul_routes.append(route)
seoul_routes = pd.concat(seoul_routes)
seoul_routes
```

### 중복된 경로 제거

```python
seoul_routes = seoul_routes.drop_duplicates()
seoul_routes
```

### 불필요한 columns 제거

```python
seoul_routes = seoul_routes[['patNo', 'date', 'reason', 'route']]
seoul_routes
```

### '확인중', '미부여',  '이동경로 공개기간 경과' 등 제거

```python
seoul_routes['patNo'] = seoul_routes['patNo'].apply(str)
seoul_routes = seoul_routes[~seoul_routes.patNo.str.contains('확인중')]
seoul_routes = seoul_routes[~seoul_routes.patNo.str.contains('미부여')]
seoul_routes = seoul_routes[~seoul_routes.route.str.contains('이동경로 공개기간 경과')]
seoul_routes = seoul_routes[~seoul_routes.route.str.contains('확인중')]
seoul_routes
```

### Date column

```python
hangeul = seoul_routes[seoul_routes['date'].str.match('\d*월 \d*일')== True]
hangeul
```

```python
hangeul['date'] = '2020년 ' + hangeul['date'].astype(str)
hangeul['date'] = pd.to_datetime(hangeul['date'], format='%Y년 %m월 %d일')
hangeul
```

```python
dash = seoul_routes[seoul_routes['date'].str.match('\d*/\d*')== True]
dash
```

```python
dash['date'] = '2020/' + dash['date'].astype(str)
dash['date'] = pd.to_datetime(dash['date'], format='%Y/%m/%d')
dash
```

```python
seoul_routes = pd.concat([hangeul, dash])
seoul_routes = seoul_routes.sort_values('date')
seoul_routes = seoul_routes.reset_index(drop=True)
seoul_routes
```

### route column

```python
date_aparted = pd.DataFrame()
for index, row in seoul_routes.iterrows():
    one_route = row['route']
    indices = []
    dates = []
    for match in re.finditer('\d*월 \d*일', one_route):
        indices.append(match.span()[0])
        date = datetime.strptime('2020년 %s' % one_route[match.start():match.end()], '%Y년 %m월 %d일')
        dates.append(date)
    parts = [one_route[i:j] for i, j in zip(indices, indices[1:]+[None])]
    for i, part in enumerate(parts):
        date_aparted = date_aparted.append({'patNo': row['patNo'], 'reason': row['reason'],
                     'date': dates[i], 'route': part}, ignore_index=True)

date_aparted
```

```python
visit_aprted = pd.DataFrame()
for index, row in date_aparted.iterrows():
    one_route = row['route']
    visits = one_route.split('→')
    for visit in visits:
        visit_aprted = visit_aprted.append({'date': row['date'], 'patNo': str(int(row['patNo'])),
                                       'reason': row['reason'], 'route': visit}, ignore_index=True)
        
visit_aprted
```

### 날짜만 있는 row 제거

```python
regex_applied = pd.DataFrame()
for index, row in visit_aprted.iterrows():
    one_route = row['route']
    one_route = re.sub('\d*월 \d*일: \d*:\d* ~ \d*:\d*', '', one_route)
    one_route = re.sub('\d*월 \d*일: \d*:\d*', '', one_route)
    one_route = re.sub('\d*월 \d*일\s*\([가-힣]\):*', '', one_route)
    one_route = re.sub('\d*월 \d*일 ~', '', one_route)
    one_route = re.sub('\d*월 \d*일', '', one_route)
    one_route = re.sub('\d*월 \d*\([가-힣]\)', '', one_route)
    one_route = re.sub('\d*:\d* ~ \d*:\d*', '', one_route)
    one_route = re.sub('\d*:\d*', '', one_route)
    one_route = re.sub('^\s*,\s*', '', one_route)
    one_route = re.sub('^\s*', '', one_route)
    one_route = re.sub('^~\s*', '', one_route)
    one_route = re.sub('^∼\s*', '', one_route)
    one_route = re.sub('^-\s*', '', one_route)
    one_route = re.sub('$\s*', '', one_route)
    one_route = re.sub('$\n*', '', one_route)
    one_route = re.sub('^\d*일\([가-힣]\)', '', one_route)
    one_route = re.sub('^\s*', '', one_route)
    if one_route == "": continue
    
    regex_applied = regex_applied.append({'patNo': row['patNo'], 'reason': row['reason'],
                     'date': row['date'], 'route': one_route}, ignore_index=True)
        
regex_applied
```
